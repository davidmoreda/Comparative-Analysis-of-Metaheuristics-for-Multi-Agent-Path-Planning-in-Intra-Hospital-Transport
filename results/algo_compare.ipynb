```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import ast
import os

plt.style.use("seaborn-v0_8")

# -------------------------------------------------------
# Funcin para cargar un .db y expandir params (si existe)
# -------------------------------------------------------
def load_db(path):
    con = sqlite3.connect(path)
    df = pd.read_sql_query("SELECT * FROM runs", con)
    con.close()
    
    if "params" in df.columns:
        try:
            df_params = df["params"].apply(ast.literal_eval).apply(pd.Series)
            df = pd.concat([df.drop(columns=["params"]), df_params], axis=1)
        except:
            print(" No puedo expandir 'params'. Est bien formateado?")
    
    return df


# -------------------------------------------------------
# Funcin para generar grficas comunes a cualquier GA
# -------------------------------------------------------
def plot_grid_results(df, title_prefix=""):
    algo_name = df["algo"].iloc[0]
    
    print(f"\n===== PLOTTING {algo_name.upper()} =====")
    print(df.describe())
    
    # Seleccin automtica de columnas de hiperparmetros numricos
    hypers = [c for c in df.columns if c not in ["id", "algo", "seed", "score", "time_sec"] 
              and df[c].dtype != "O"]
    
    print("\nHiperparmetros detectados:", hypers)

    # Color por seed o por algo interesante
    colors = df["seed"] if "seed" in df else None

    # -------------------------
    # Subplots dinmicos
    # -------------------------
    n = len(hypers)
    fig, axes = plt.subplots(1, n, figsize=(5*n, 5))

    if n == 1: 
        axes = [axes]

    for ax, h in zip(axes, hypers):
        ax.scatter(df[h], df["score"], c=colors, cmap="viridis", alpha=0.7)
        ax.set_xlabel(h)
        ax.set_ylabel("Score")
        ax.set_title(f"{title_prefix}{algo_name}: Score vs {h}")

    plt.tight_layout()
    plt.show()

    # -------------------------
    # Tiempo vs Score
    # -------------------------
    plt.figure(figsize=(6,5))
    plt.scatter(df["time_sec"], df["score"],
                c=df["seed"], cmap="viridis", alpha=0.7)
    plt.xlabel("Runtime (s)")
    plt.ylabel("Score")
    plt.title(f"{title_prefix}{algo_name}: Runtime vs Score")
    plt.show()


# -------------------------------------------------------
# Cargar todos los .db automticamente en la carpeta actual
# -------------------------------------------------------

db_files = [
    "grid_results_ga.db",
    "grid_results_mulambda.db",
    "grid_results_sa.db",
    "nsga2_grid.db"
]

for db in db_files:
    if os.path.exists(db):
        df = load_db(db)
        plot_grid_results(df, title_prefix=f"[{db}] ")

```

```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import ast
import os

plt.style.use("seaborn-v0_8")


# ==========================================================
# LOAD DB
# ==========================================================
def load_db(path):
    
    con = sqlite3.connect(path)
    df = pd.read_sql_query("SELECT * FROM runs", con)
    con.close()
    
    # NSGA-II NO tiene params  skip
    if "params" in df.columns:
        try:
            df_params = df["params"].apply(ast.literal_eval).apply(pd.Series)
            df = pd.concat([df.drop(columns=["params"]), df_params], axis=1)
        except:
            pass

    return df


# ==========================================================
# PLOTS para GA, MU+LAMBDA, SA
# ==========================================================
def plot_classic(df, db_name=""):

    algo_name = db_name.replace(".db","")

    # columnas que NO son hyperparmetros
    exclude = ["id", "seed", "score", "time_sec", "algo"]

    hypers = [c for c in df.columns 
              if c not in exclude 
              and pd.api.types.is_numeric_dtype(df[c])]

    print(f"\n==== {algo_name.upper()} detected ====")
    print("Hypers:", hypers)

    # ---- Subplots por hyper ---
    fig, axes = plt.subplots(1, len(hypers), figsize=(5*len(hypers), 5))
    if len(hypers) == 1:
        axes = [axes]

    for ax, h in zip(axes, hypers):
        ax.scatter(df[h], df["score"],
                   c=df["seed"], cmap="viridis", alpha=0.7)
        ax.set_xlabel(h)
        ax.set_ylabel("Score")
        ax.set_title(f"{algo_name}: Score vs {h}")

    plt.tight_layout()
    plt.show()

    # ---- Tiempo vs Score ----
    plt.figure(figsize=(6,5))
    plt.scatter(df["time_sec"], df["score"],
                c=df["seed"], cmap="viridis", alpha=0.7)
    plt.xlabel("Runtime (s)")
    plt.ylabel("Score")
    plt.title(f"{algo_name}: Runtime vs Score")
    plt.show()



# ==========================================================
# PLOTS para NSGA-II
# ==========================================================
def plot_nsga(df):

    print("\n==== NSGA-II detected ====")

    # hiperparmetros reales
    hypers = ["pop_size", "ngen", "cxpb", "mutpb"]

    # ---- Score multiobjetivo ----
    plt.figure(figsize=(6,5))
    plt.scatter(df["clean"], df["penalized"],
                c=df["conflicts"], cmap="plasma", alpha=0.8)
    plt.xlabel("Clean objective")
    plt.ylabel("Penalized objective")
    plt.title("NSGA-II: Clean vs Penalized")
    plt.colorbar(label="Conflicts")
    plt.show()

    # ---- HV vs hypers ----
    fig, axes = plt.subplots(1, len(hypers), figsize=(5*len(hypers), 5))

    for ax, h in zip(axes, hypers):
        ax.scatter(df[h], df["hv"], c=df["seed"], cmap="viridis", alpha=0.7)
        ax.set_xlabel(h)
        ax.set_ylabel("Hypervolume")
        ax.set_title(f"HV vs {h}")

    plt.tight_layout()
    plt.show()

    # ---- Tiempo vs HV ----
    plt.figure(figsize=(6,5))
    plt.scatter(df["time_sec"], df["hv"],
                c=df["seed"], cmap="viridis", alpha=0.8)
    plt.xlabel("Runtime (s)")
    plt.ylabel("Hypervolume")
    plt.title("NSGA-II: Runtime vs HV")
    plt.show()


# ==========================================================
# Ejecutar para todos los DB
# ==========================================================
db_files = [
    "grid_results_ga.db",
    "grid_results_mulambda.db",
    "grid_results_sa.db",
    "nsga2_grid.db"
]

for db in db_files:
    if os.path.exists(db):
        df = load_db(db)

        # NSGA-II
        if "hv" in df.columns:
            plot_nsga(df)
        # GA, MU+LAMBDA, SA
        else:
            plot_classic(df, db_name=db)
```
